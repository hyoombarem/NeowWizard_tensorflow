### 로지스틱 회귀

분류 (Classification)
- 여려 변수를 입력받고 종합하여 0, 1 로 분류해내는 회귀분석
- Sigmoid 함수를 이용하여 결과를 0과 1 사이의 값으로 나타내어 확률처럼 사용한다
- 최종으로 Cross Entropy를 적용하여 loss값을 계산한다

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD, Adam

import numpy as np
```

```python
loaded_data = np.loadtxt("diabetes.csv", delimiter = ',')
    
x_data = loaded_data[ :, 0:-1]
y_data = loaded_data[ :, [-1]]

print("x_data.shape = ", x_data.shape)
print("y_data.shape = ", y_data.shape)
```

```python
# 모델 구축

model = Sequential()

model.add(Dense(y_data.shape[1], input_shape = (x_data.shape[1], ), activation = 'sigmoid'))
```

```python
# 모델 컴파일
model.compile(optimizer = SGD(learning_rate = 0.01), loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# 모델 확인

model.summary()
```

```python
# 모델 학습

hist = model.fit(x_data, y_data, epochs = 500, validation_split = 0.2, verbose = 2)
```

```python
model.evaluate(x_data, y_data)
```

```python
# 손실 모델의 정확도
# 학습의 진행도에 따른 손실값(loss)의 변화
import matplotlib.pyplot as plt

plt.title("Loss")
plt.xlabel('epochs')
plt.ylabel('loss')
plt.grid()

plt.plot(hist.history["loss"], label = "train loss")
plt.plot(hist.history["val_loss"], label = "validation loss")

plt.legend(loc = "best")

plt.show()
```

```python
# 학습이 진행됨에 따라 정확도(Accuracy)의 변화

plt.title("Accuracy")
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.grid()

plt.plot(hist.history['accuracy'], label = 'train accuracy')
plt.plot(hist.history["val_accuracy"], label = 'validation accuracy')

plt.legend(loc = "best")

plt.show()
```
