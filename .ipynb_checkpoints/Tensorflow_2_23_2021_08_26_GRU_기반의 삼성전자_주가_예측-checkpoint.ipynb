```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout
```

```python
# 데이터 로드
raw_df = pd.read_csv("./005930.KS.csv")

raw_df.head()
```

```python
# 데이터 로드 및 분포확인

plt.figure(figsize = (7, 4))

plt.title('SAMSUNG ELECTRIC STOCK PRICE')
plt.ylabel('price (won)')
plt.xlabel('period (day)')
plt.grid()

plt.plot(raw_df["Adj Close"], label = "Adj Close", color = "blue")
plt.legend(loc = "best")

plt.show()

# GRU를 이용하여 전반부 데이터 학습
```

```python
raw_df.isnull().sum()
```

```python
raw_df.loc[raw_df["Open"].isna()]
```

```python
# 결측치 탐색
# Volume이 0인 데이터는 그냘 주식 시장이 안열어서 데이터가 없는 경우

raw_df["Volume"] = raw_df["Volume"].replace(0, np.nan)

for col in raw_df.columns:
    missing_rows = raw_df.loc[raw_df[col] == 0].shape[0]
    print(col + ": " + str(missing_rows))
```

```python
raw_df.isnull().sum()
```

```python
raw_df = raw_df.dropna()
raw_df.isnull().sum()
```

```python
# 데이터 정규화

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scale_cols = ["Open", "High", "Low", "Close", "Adj Close", "3MA", "5MA", "Volume"]

scaled_df = scaler.fit_transform(raw_df[scale_cols])
print(type(scaled_df), "\n")
scaled_df = pd.DataFrame(scaled_df, columns = scale_cols)
print(scaled_df)
```

```python
feature_cols = ["3MA", "5MA", "Adj Close"]
label_cols = ["Adj Close"]

label_df = pd.DataFrame(scaled_df, columns = label_cols)
feature_df = pd.DataFrame(scaled_df, columns = feature_cols)

print(feature_df)
print(label_df)

label_np = label_df.to_numpy()
feature_np = feature_df.to_numpy()
```

```python
# 시계열 데이터 생성 및 모델 구축

window_size = 40

X, Y = make_sequence_dataset(feature_np, label_np, window_size)

print(X.shape, Y.shape)
```

```python
def make_sequence_dataset(feature, label, window_size):
    feature_list = []
    label_list = []
    
    for i in range(len(feature) - window_size):
        feature_list.append(feature[i:i+window_size])
        label_list.append(label[i + window_size])
        
    return np.array(feature_list), np.array(label_list)
```

```python
# 모델 생성

split = -200
X_train = X[0:split]
Y_train = Y[0:split]

X_test = X[split:]
Y_test = Y[split:]

print(X_train.shape, Y_train.shape)
print(X_test.shape, Y_test.shape)
print(X_train[:6, ])
```

```python
model = Sequential()

model.add(GRU(256,
             activation = 'tanh', 
             input_shape = X_train[0].shape))
model.add(Dense(1, activation = "linear"))
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping

model.compile(loss = "mse", optimizer = "adam", metrics = ["mae"])

early_stop = EarlyStopping(monitor = 'val_loss', patience = 5)

model.fit(X_train, Y_train, validation_data = (X_test, Y_test), epochs = 100, batch_size = 16, callbacks = [early_stop])
```

```python
pred = model.predict(X_test)

plt.figure(figsize = (12, 6))
plt.title("3MA + 5MA + Adj Close, window_size = 40")
plt.ylabel('adj close')
plt.xlabel('period')
plt.plot(Y_test, label = 'actual')
plt.plot(pred, label = 'prediction')
plt.grid()
plt.legend(loc = 'best')

plt.show()
```
