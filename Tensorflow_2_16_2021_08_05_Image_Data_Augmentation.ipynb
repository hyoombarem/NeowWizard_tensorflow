### 데이터 어그맨테이션
- Data Augmentation
- 원본 이미지에 좌우반전, 이동, 회전 등을 이용하여 비슷한 새로운 데이터를 많들어 양질의 데이터를 제작


```python
# ImageDataGenerator(rotatoin_range = 30, horizontal_flip = True, rescale = 1/255)
# load_img 로 불러오면 데이터 타입니 jpegimageFile 형식으로 반횐됨

```

```python
# flow() 함수 예제

import tensorflow as tf

import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.preprocessing.image import load_img, img_to_array, ImageDataGenerator
```

```python
# 이미지 생성

gen = ImageDataGenerator(rotation_range = 30, 
                        width_shift_range = 0.3, 
                        shear_range = 0.4, 
                        horizontal_flip = True)
```

```python
# 이미지 로드 및 정규화

img_array_list = []
img_names = ['./dog1.jpg', './dog2.jpg',
            './cat1.jpg', './cat2.jpg']

for i in range(len(img_names)):
    
    loaded_img = load_img(img_names[i], target_size = (100, 100))
    
    loaded_img_array = img_to_array(loaded_img) / 255.0
    
    img_array_list.append(loaded_img_array)
```

```python
# 사진 송출

plt.figure(figsize = (6, 6))
for i in range(len(img_array_list)):
    plt.subplot(1, len(img_array_list), i+1)
    plt.xticks([])
    plt.yticks([])
    plt.title(img_names[i])
    plt.imshow(img_array_list[i])
```

```python
# flow() 함수의 적용

batch_size = 2

data_gen = gen.flow(np.array(img_array_list), 
                   batch_size = batch_size)
```

```python
# next() 함수 실행 및 변형 이미지 출력

img = data_gen.next()

plt.figure(figsize = (6, 6))
for i in range(len(img)):
    plt.subplot(1, len(img), i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img[i])
```

```python
# flow_from-daic

import tensorflow as tf

import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.preprocessing.image import load_img, img_to_array, ImageDataGenerator
```

```python
gen = ImageDataGenerator(rotation_range = 30, 
                        width_shift_range = 0.3, 
                        shear_range = 0.4, 
                        rescale = 1./255)
```

```python
# flow_from_directory( 적용)

data_path = './test_dir/'
batch_size = 3
data_gen = gen.flow_from_directory(directory = data_path, 
                                  batch_size = batch_size, 
                                  shuffle = True, 
                                  target_size = (100, 100), 
                                  class_mode = 'categorical')
```

```python
# 변환한 이미지 생성

img, label = data_gen.next()

plt.figure(figsize = (6, 6))
for i in range(len(img)):
    plt.subplot(1, len(img), i+1)
    plt.xticks([])
    plt.yticks([])
    
    plt.title(str(np.argmax(label[i])))
    
    plt.imshow(img[i])
```
